```python
from Models import MultiLayerNN, OneLayerNN
import torchinfo, torch, Data
import numpy as np
import matplotlib.pyplot as plt 
from tqdm import tqdm 
from IPython.display import display, clear_output
import time
from Utility import getFourierCoeffs

# if torch.cuda.is_available():
#     device = torch.device("cuda")
# else:
device = torch.device("cpu")

print("Using device", device)
```

```python
def true_func(x):
    return np.sin(np.pi*x)

def func1(x):
    noise = 0.05*np.sin(3*np.pi*x) + 0.4*np.sin(8*np.pi*x) + 0.23*np.sin(16*np.pi*x) 
    return true_func(x) + noise

def add_noise_func(x, p):
    vals = true_func(x)
    for i in range(len(vals)):
        if np.random.random() <= p:
            vals[i] = vals[i] + np.random.normal(loc=0, scale=0.25)
    return vals

np.random.seed(33)
x_train = np.linspace(-1,1,2000)
y_train = func1(x_train)
xy_ds = Data.FuncData(x_train, y_train)
xy_dl = torch.utils.data.DataLoader(xy_ds, batch_size=50, shuffle=True)
```

```python
plt.plot(x_train, y_train)
plt.title("Noisy Function")
plt.xlabel("$x$")

# Compute benchmark error
BENCHMARK_MSE = np.mean(np.square(y_train - true_func(x_train)))
print("BENCHMARK MSE =", BENCHMARK_MSE)
```

```python
def compute_NN_function(num_epochs, hidden_units, xy_ds, xy_dl):
    def weights_init(m):
        if isinstance(m, torch.nn.Linear):
            torch.nn.init.normal_(m.weight, mean=0.0, std=1.0)
            torch.nn.init.zeros_(m.bias)

    # define init method inside your model class
    def init_with_normal(model):
        model.apply(weights_init)
        
    def getFourierCoeffs(x, func, num=11):
        num_pts = len(x)
        sin_coeffs = []
        cos_coeffs = []
        for i in range(1,num+1):
            sin_coeffs.append(sum(func(x) * np.sin(i*np.pi*x)) * ((np.max(x)-np.min(x))/num_pts))
            cos_coeffs.append(sum(func(x) * np.cos(i*np.pi*x)) * ((np.max(x)-np.min(x))/num_pts))
        zero_cos_coeff = sum(func(x)) * ((np.max(x)-np.min(x))/num_pts)
        return (sin_coeffs, [zero_cos_coeff] + cos_coeffs)

    np.random.seed(1113)
    torch.manual_seed(1113)
    x_test = torch.tensor(2*np.random.random(size=25) - 1).float().unsqueeze(-1).to(device)
    # model = MultiLayerNN([32,32]).to(device)
    # NOTE: What is the effect of increasing the number of layers 
    # on the frequency components of the learned function?
    model = OneLayerNN(hidden_size=hidden_units).to(device)
    init_with_normal(model)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.002 * 32/hidden_units)
    loss = torch.nn.MSELoss(reduction='sum')
    NUM_EPOCH = num_epochs
    CONVERGE_EPS = 0.5*BENCHMARK_MSE
    print("CONVERGING LOSS = ", CONVERGE_EPS)
    STOP_EPOCH = num_epochs
    curr_loss = np.inf
    torch.cuda.empty_cache()
    epoch = 0
    sin_coeffs_mat = []
    func_hat = lambda x: model(torch.tensor(x).to(device).float().unsqueeze(-1)).detach().numpy().flatten()
    while curr_loss > CONVERGE_EPS:
        if epoch>=STOP_EPOCH:
            print("Model didn't converge to desired loss value", CONVERGE_EPS)
            break
        epoch_loss = 0
        epoch += 1
        for X,Y in xy_dl:
            _X = X.to(device)
            _Y = Y.to(device)
            Y_hat = model(_X)
            optimizer.zero_grad()
            L = loss(_Y, Y_hat) / len(xy_dl.dataset)
            L.backward()
            optimizer.step()
            epoch_loss += float(L.item())
        curr_loss = epoch_loss
        sin_coeffs, _ = getFourierCoeffs(x_train, func_hat, num=12)
        sin_coeffs_mat.append(sin_coeffs)
        print('Epoch',epoch,'| Loss:', str(np.round(epoch_loss,6)), end='\r')

    y_test = model(x_test)
    x_test = x_test.cpu().detach().numpy()
    y_test = y_test.cpu().detach().numpy()
    
    return x_test, y_test, sin_coeffs_mat
```

```python
x_test, y_test, sin_coeffs_mat = compute_NN_function(300, 32, xy_ds, xy_dl)
x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
x_test = x_test.flatten()
y_test = y_test.flatten()
plt.scatter(x_test, y_test, label='NN Function', marker='x')
plt.scatter(x_test, true_func(x_test), label='True function')
sort_idx = np.argsort(x_test.flatten())
plt.plot(x_test[sort_idx], func1(x_test)[sort_idx], label='Noisy Function', alpha=0.3)
plt.legend()
plt.xlabel("$x$")
plt.show()

plt.figure()
sin_coeffs_mat = np.array(sin_coeffs_mat)
for i in range(sin_coeffs_mat.shape[1]):
    if i==1-1 or i==8-1:
        plt.plot(sin_coeffs_mat[:,i])
```

```python

hidden_sizes = list(range(100, 5000, 50))
test_errors = []
for i in tqdm(range(len(hidden_sizes))):
    s = hidden_sizes[i]
    x_test, y_hat = compute_NN_function(-1, s, xy_ds, xy_dl)
    y_true = true_func(x_test)
    test_error = np.mean(np.square(y_hat-y_true))
    test_errors.append(test_error)
```

```python
plt.plot(hidden_sizes, test_errors)
plt.xlabel("Num. of Hidden Units")
```
